Summit County, CO spans the Continental Divide and is home to ski resorts, high mountain towns, I-70, and wilderness within the National Forest System. It is part of the ancestral home of the Ute people.

```python
# Define the URL for Los Angeles mean temperature data download
summco_temp_url = (
    'https://www.ncei.noaa.gov/access/monitoring/climate-at-a-glance/county/'
    'time-series/CO-117/tmax/1/8/1896-2023.csv?base_prd=true&begbaseyear=1901&'
    'endbaseyear=2000)')
```

```python
# Import LA temperature values from NCEI
summco_temp_df = pd.read_csv(
    summco_temp_url, header=4, names=['year', 'temperature_f','anomaly'])
```

```python
# Extract the year from the date
summco_temp_df.year = pd.to_datetime(summco_temp_df.year, format='%Y%m').dt.year  
```

```python
# Compute Celcius temperatures
summco_temp_df['temperature_c'] = (summco_temp_df['temperature_f'] - 32) * 5 / 9
```

```python
# Plot the Summit County temperature time series
summco_temp_df.plot(
    x='year', y='temperature_f',
    title='August Mean Daily Maximum Temperature in Summit County, CO, USA',
    xlabel='Year', ylabel='Temperature (F)');
```

Mean daily maximum temperatures for the month of August from 1896 - 2023. From NCEI: "maximum and minimum temperature data are derived from a 5km gridded instance of the nClimGrid dataset (Vose et al.2014) using data from the GHCN-Daily database."
